Processed Folder Creation:

```python
import os
import librosa
import soundfile as sf
import numpy as np

TARGET_SR = 16000
MAX_LEN = TARGET_SR * 3  # 3 seconds
SUPPORTED_EXT = (".wav", ".mp3")

def preprocess(in_path, out_path):
    y, _ = librosa.load(in_path, sr=TARGET_SR, mono=True)
    y, _ = librosa.effects.trim(y, top_db=25)

    if len(y) > MAX_LEN:
        y = y[:MAX_LEN]
    else:
        y = np.pad(y, (0, MAX_LEN - len(y)))

    y = y / (np.max(np.abs(y)) + 1e-9)
    sf.write(out_path, y, TARGET_SR)


def process_folder(src, dst, prefix):
    os.makedirs(dst, exist_ok=True)

    for f in os.listdir(src):
        if f.lower().endswith(SUPPORTED_EXT):
            in_path = os.path.join(src, f)

            base = os.path.splitext(f)[0]
            out_file = f"{prefix}_{base}.wav"
            out_path = os.path.join(dst, out_file)

            preprocess(in_path, out_path)


# -------- AI VOICES --------
for lang in ["english", "hindi", "tamil", "telugu", "malayalam"]:
    for gender in ["female", "male"]:
        src = f"ai voice/{lang}/{lang}_{gender}"
        dst = f"processed/ai/{lang}"

        if os.path.exists(src):
            process_folder(src, dst, prefix=f"ai_{lang}_{gender}")


# -------- HUMAN VOICES --------
human_map = {
    "english": ["english_human"],
    "hindi": ["hindi_human"],
    "tamil": ["tamil_human"],
    "telugu": ["telugu_human"],
    "malayalam": ["malayalam_human"],
}

for lang, folders in human_map.items():
    for folder in folders:
        src = f"human voice/{folder}"
        dst = f"processed/human/{lang}"

        if os.path.exists(src):
            process_folder(src, dst, prefix=f"human_{lang}")


print("ALL AUDIO (WAV + MP3) PREPROCESSED")
```

```python
import os
import librosa
import soundfile as sf
import numpy as np

TARGET_SR = 16000
MAX_LEN = TARGET_SR * 3  # 3 seconds
SUPPORTED_EXT = (".wav", ".mp3")

def preprocess_file(in_path, out_path):
    y, _ = librosa.load(in_path, sr=TARGET_SR, mono=True)
    y, _ = librosa.effects.trim(y, top_db=25)

    if len(y) < MAX_LEN:
        return False  # skipped

    if len(y) > MAX_LEN:
        y = y[:MAX_LEN]
    else:
        y = np.pad(y, (0, MAX_LEN - len(y)))

    y = y / (np.max(np.abs(y)) + 1e-9)
    sf.write(out_path, y, TARGET_SR)
    return True
```

```python
# -------- ENGLISH AI --------
ai_src = "ai voice/english"
ai_dst = "processed/ai/english"
os.makedirs(ai_dst, exist_ok=True)

for f in os.listdir(ai_src):
    if f.lower().endswith(SUPPORTED_EXT):
        out = os.path.join(ai_dst, f"ai_english_{os.path.splitext(f)[0]}.wav")
        if not os.path.exists(out):
            preprocess_file(os.path.join(ai_src, f), out)

# -------- ENGLISH HUMAN --------
human_src = "human voice/english_human"
human_dst = "processed/human/english"
os.makedirs(human_dst, exist_ok=True)

for f in os.listdir(human_src):
    if f.lower().endswith(SUPPORTED_EXT):
        out = os.path.join(human_dst, f"human_english_{os.path.splitext(f)[0]}.wav")
        if not os.path.exists(out):
            preprocess_file(os.path.join(human_src, f), out)

print(" English AI & Human processed")
```
