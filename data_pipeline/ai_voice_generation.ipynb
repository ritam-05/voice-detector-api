```python
import requests
import os
import time

# ==============================
# 🔐 AZURE CREDENTIALS (EDIT)
# ==============================
AZURE_SPEECH_KEY = "key"
AZURE_REGION = "centralindia"   # example: centralindia

TTS_URL = f"https://{AZURE_REGION}.tts.speech.microsoft.com/cognitiveservices/v1"

HEADERS = {
    "Ocp-Apim-Subscription-Key": AZURE_SPEECH_KEY,
    "Content-Type": "application/ssml+xml",
    "X-Microsoft-OutputFormat": "riff-16khz-16bit-mono-pcm",
    "User-Agent": "ai-voice-detector"
}

os.makedirs("dataset/ai/hindi", exist_ok=True)
print("Azure config ready")
```

```python
ssml = """
<speak version="1.0" xml:lang="hi-IN">
    <voice name="hi-IN-SwaraNeural">
        यह एक उच्च गुणवत्ता वाली हिंदी एआई आवाज़ है।
    </voice>
</speak>
"""

response = requests.post(TTS_URL, headers=HEADERS, data=ssml.encode("utf-8"))

with open("test_hindi_azure.wav", "wb") as f:
    f.write(response.content)

from IPython.display import Audio
Audio("test_hindi_azure.wav")
```

```python
with open("sentences_hindi.txt", "r", encoding="utf-8") as f:
    sentences = [line.strip() for line in f if line.strip()]

print("Total sentences:", len(sentences))
```

```python
for i, text in enumerate(sentences):
    print(f"Generating {i+1}/{len(sentences)}")

    ssml = f"""
    <speak version="1.0" xml:lang="hi-IN">
        <voice name="hi-IN-SwaraNeural">
            {text}
        </voice>
    </speak>
    """

    response = requests.post(
        TTS_URL,
        headers=HEADERS,
        data=ssml.encode("utf-8")
    )

    if response.status_code != 200:
        print("❌ Error:", response.status_code, response.text)
        break

    output_path = f"dataset/ai/hindi/hindi_ai_{i:03d}.wav"
    with open(output_path, "wb") as f:
        f.write(response.content)

    time.sleep(0.15)  # throttling safety

print("✅ DONE: Hindi AI voices generated with Azure")
```

Tamil AI Voice Generation:

```python
import random

OUTPUT_FILE = "sentences_tamil.txt"
NUM_SENTENCES = 500

subjects = [
    "இந்த அமைப்பு",
    "இந்த சேவை",
    "இந்த பயன்பாடு",
    "முறைமை",
    "தளம்"
]

objects = [
    "உங்கள் கோரிக்கையை",
    "உங்கள் தகவலை",
    "உங்கள் விவரங்களை",
    "தரவுகளை",
    "பயனர் சுயவிவரத்தை",
    "பதிவுகளை"
]

verbs = [
    "செயலாக்குகிறது",
    "சரிபார்க்கிறது",
    "புதுப்பிக்கிறது",
    "ஆய்வு செய்கிறது",
    "மதிப்பாய்வு செய்கிறது",
    "கையாளுகிறது"
]

endings = [
    ".",
    ". தயவுசெய்து காத்திருக்கவும்.",
    ". செயல்முறை நடைபெற்று வருகிறது.",
    ". இது சில நேரம் எடுக்கலாம்.",
    ". விரைவில் முடிக்கப்படும்."
]

sentences = set()

while len(sentences) < NUM_SENTENCES:
    sentence = (
        random.choice(subjects) + " " +
        random.choice(objects) + " " +
        random.choice(verbs) +
        random.choice(endings)
    )
    sentences.add(sentence)

with open(OUTPUT_FILE, "w", encoding="utf-8") as f:
    for s in sentences:
        f.write(s + "\n")

print(f"✅ Generated {len(sentences)} Tamil sentences → {OUTPUT_FILE}")
```

```python
import random

OUTPUT_FILE = "sentences_english.txt"
NUM_SENTENCES = 500

subjects = [
    "The system",
    "This service",
    "This application",
    "The platform",
    "The software"
]

objects = [
    "your request",
    "your information",
    "your data",
    "your profile",
    "the records",
    "the details"
]

verbs = [
    "is processing",
    "is verifying",
    "is updating",
    "is analyzing",
    "is reviewing",
    "is handling"
]

endings = [
    ".",
    ". Please wait.",
    ". The process is ongoing.",
    ". This may take some time.",
    ". It will be completed shortly."
]

sentences = set()

while len(sentences) < NUM_SENTENCES:
    sentence = (
        random.choice(subjects) + " " +
        random.choice(objects) + " " +
        random.choice(verbs) +
        random.choice(endings)
    )
    sentences.add(sentence)

with open(OUTPUT_FILE, "w", encoding="utf-8") as f:
    for s in sentences:
        f.write(s + "\n")

print(f"✅ Generated {len(sentences)} English sentences → {OUTPUT_FILE}")
```

```python
import requests
import os
import time

# 🔐 AZURE DETAILS
AZURE_SPEECH_KEY = "key"
AZURE_REGION = "centralindia"


TTS_URL = f"https://{AZURE_REGION}.tts.speech.microsoft.com/cognitiveservices/v1"

HEADERS = {
    "Ocp-Apim-Subscription-Key": AZURE_SPEECH_KEY,
    "Content-Type": "application/ssml+xml",
    "X-Microsoft-OutputFormat": "riff-16khz-16bit-mono-pcm",
    "User-Agent": "ai-voice-detector"
}

os.makedirs("dataset/ai/english_female", exist_ok=True)
os.makedirs("dataset/ai/english_male", exist_ok=True)

print("Azure English TTS ready")
```

```python
ssml = """
<speak version="1.0" xml:lang="en-IN">
    <voice name="en-IN-NeerjaNeural">
        This is a test of a female English AI voice with an Indian accent.
    </voice>
</speak>
"""

r = requests.post(TTS_URL, headers=HEADERS, data=ssml.encode("utf-8"))

with open("test_english_female.wav", "wb") as f:
    f.write(r.content)

from IPython.display import Audio
Audio("test_english_female.wav")
```

```python
with open("sentences_english.txt", "r", encoding="utf-8") as f:
    sentences = [line.strip() for line in f if line.strip()]

print("Total English sentences:", len(sentences))
```

```python
for i, text in enumerate(sentences[:500]):
    print(f"Female {i+1}/500")

    ssml = f"""
    <speak version="1.0" xml:lang="en-IN">
        <voice name="en-IN-NeerjaNeural">
            {text}
        </voice>
    </speak>
    """

    r = requests.post(TTS_URL, headers=HEADERS, data=ssml.encode("utf-8"))

    if r.status_code != 200:
        print("❌ Error:", r.status_code, r.text)
        break

    with open(f"dataset/ai/english_female/english_female_{i:03d}.wav", "wb") as f:
        f.write(r.content)

    time.sleep(0.15)

print("✅ 500 English FEMALE voices generated")
```

```python
for i, text in enumerate(sentences[:500]):
    print(f"Male {i+1}/500")

    ssml = f"""
    <speak version="1.0" xml:lang="en-IN">
        <voice name="en-IN-PrabhatNeural">
            {text}
        </voice>
    </speak>
    """

    r = requests.post(TTS_URL, headers=HEADERS, data=ssml.encode("utf-8"))

    if r.status_code != 200:
        print("❌ Error:", r.status_code, r.text)
        break

    with open(f"dataset/ai/english_male/english_male_{i:03d}.wav", "wb") as f:
        f.write(r.content)

    time.sleep(0.15)

print("✅ 500 English MALE voices generated")
```
